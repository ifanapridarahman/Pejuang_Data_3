# Lambda

Fungsi lambda adalah fungsi anonim **one line** yang didefinisikan tanpa nama. Fungsi lambda dapat mengambil sejumlah **argument**, tetapi hanya dapat memiliki satu **expression**. Sementara fungsi normal didefinisikan menggunakan kata kunci **def**, dalam Python fungsi anonim didefinisikan menggunakan kata kunci lambda.

Fungsi Lambda digunakan ketika fungsi sederhana hanya digunakan sekali atau untuk waktu yang singkat dalam kode kita. Penggunaannya yang paling umum adalah sebagai argumen untuk fungsi tingkat tinggi (hanya yang menggunakan fungsi lain sebagai argumen

Lambda expression di Python adalah sebuah ekspressi untuk membuat fungsi

Lambda sendiri berasal dari teori kalkulus, yakni Lambda Calculus yang dikenalkan oleh Alonzo Chrurch di tahun 1930

Berkat lambda kita bisa membuat fungsi tanpa nama atau dikenal juga dengan anonymous function.

Misalkan kita mempunyai fungsi penjumlahan kuadrat seperti ini:


```python
def suqare_sum(x,y):
  return x**2 + y**2
```

dengan lambda.. kita tidak perlu menggunakan def dan return, bahkan kita juga tidak perlu menuliskan namanya

```python
lambda x,y : x**2 + y**2
```

# Menggunakan Lambda Expression

Intinya, lambda digunakan untuk membuat fungsi dalam satu baris ekspresi.

ini adalah format untuk membuat fungsi lambda:

lambda-expression.png

Lambda bisa memiliki lebih dari satu argumen atau parameter, tapi hanya bisa memiliki satu ekspresi atau isi

```python
greeting = lambda name : print(f"Hello, {name}")
```

Karena fungsi lambda tidak memiliki nama, jadi kita butuh variabel untuk menyimpannya.

Nanti saat kita mau panggil, kita tinggal tuliskan saja nama variabel seperti ini:

```python
greeting("Dian")
greeting("Ayu")
```

Beberapa contoh penggunaan fungsi lambda

```python
# Fungsi lambda untuk menambahkan angka 10 kepada argument input:
f = lambda x: x+10
val1 = f(5)
val2 = f(100)
print(val1,val2)
```

```python
# Fungsi lambda untuk mengalikan dua argument input dan menampilkan hasilnya:
f = lambda x,y: x*y
val3 = f(2,10)
val4 = f(7,5)
print(val3,val4)
```

# Anonymous Function

Karena lambda adalah anonymous function, ia akan bebas menggunakan nama apa saja. Dengan kata lain, fungsi lambda bisa disimpan di variabel manapun. Ini tentunya tidak bisa dilakukan oleh fungsi yang dibuat dengan def. Coba perhatikan ini:

```python
def say_hello(nama):
  print("Hello {nama}, apa kabar?")

ucapin = say_hello(nama)
```

Saya mencoba membuat fungsi say_hello() dengan def, kemudian menyimpannya ke dalam variabel ucapin.

Tapi malah error!

Ini karena fungsi tersebut bukanlah fungsi anonymous yang boleh diubah-ubah namanya.

Lalu coba perhatikan di baris terakhir, fungis lambda greeting() yang sudah dibuat disimpan lagi ke variabel yang berbeda, yakni sapa.

# Eksekusi Lambda Secara Langsung

Eksekusi Lambda secara langsung kelebihan fungsi lambda dibandingkan def adalah bisa kita eksekusi langsung

Contohnya:

```python
(lambda x,y: x**2 + y**2)(4,6)
```

Tanda kurung yang mengapit fungsi lambda artinya kita akan langsung mengeksekusi fungsi tersebut.

Lalu kurung berikutnya akan berisi parameter

Angka 4 dan 6 adalah parameter x dan y yang akan diberikan kepada fungsi lambda.

Maka kita akan langsung mendapatkan hasilnya, yakni 42 + 63 = 52

Hasil dari eksekusi ini bisa juga kita simpan ke dalam variabel

```python
hasil = (lambda x,y: x**2 + y**2)(4,6)
```

## Mengapa Harus Pakai Lambda?

Kita memang tidak harus selalu menggunakan Lambda, tapi dalam kasus tertentu lambda lebih baik dibandingkan fungsi biasa.

Lambda biasanya dibutuhkan saat kita ingin membuat fungsi dalam satu baris

Biasanya saat menggunakan fungsi-fungsi seperti **filter(), map()** dan **reduce()** kita akan membutuhkan lambda

Mengapa?

Karena di fungsi-fungsi tersebut membutuhkan parameter fungsi

Contoh:

```python
bilangan = [10,2,8,7,5,4,3,11,0,1]
filtered_result = map(lambda x: x*x, bilangan)
print(list(filtered_result))
```

Pada fungsi map() kita memberikan parameter dengan fungsi lambda.

Mengapa bisa begitu?

Karena lambda bisa disimpan ke dalam variabek, otomatis dia akan bisa menjadi parameter

Satu lagi contoh dengan fungsi filter():


```python
# Menentukan bilangan genap
genap = lambda x: x%2 == 0
list(filter(genap, range(11)))
```
